sigmoid_function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
  return 1/(1+np.exp(-z))

z = np.linspace(-10,10,100)
plt.plot(z,sigmoid(z))
plt.show()
```

```python
x = np.linspace(-10, 10, 200)

# Example 1: w=1, b=0
plt.plot(x, sigmoid(1*x + 0), label='w=1, b=0')

# Example 2: w=2, b=-2
plt.plot(x, sigmoid(2*x - 2), label='w=2, b=-2')

# Example 3: w=-1, b=1
plt.plot(x, sigmoid(-1*x + 1), label='w=-1, b=1')

plt.title("Effect of Weights and Bias on Sigmoid")
plt.xlabel("x")
plt.ylabel("Ïƒ(z)")
plt.legend()
plt.grid()
plt.show()
```

odds function

```python
p = 0.8 # p is probability

def odds(p):
  if p<=0 or p>=1:
    return print("p should be between 0 and 1")
  return p/(1-p)

print(odds(p))
```
