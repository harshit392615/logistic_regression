Implementing Logistic regression from scrtch

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.randint(-5,5,size=(100,5))
y = (x[:,1]>0).astype(int)
```

```python
for i in range(5):
  mean = np.mean(x[: , i])
  std = np.std(x[: , i])
  x[:,i] = (x[:,i] - mean)/std
```

```python
ones = np.ones(x.shape[0])
x = np.column_stack((ones , x))
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))

def log_loss(y,p):
  esp = 1e-9
  p = np.clip(p,esp,1-esp)
  return -np.mean(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
def train(epochs=10000,lr=1):
  w = np.zeros(x.shape[1])
  losses=[]

  for i in range(epochs):
    z = x @ w
    p = sigmoid(z)

    loss = log_loss(y,p)
    grad = x.T@(p-y)/len(y)

    w -= lr*grad
    losses.append(loss)

  return w , loss
```

```python
print(train())
```
